# Realtime Factor Analysis

Data was gathered using mujoco 3.2.3 testspeed utility with the engine plugins disabled, as they are not supported for testspeed.

They are averages of 10000 steps.


```python
import pandas as pd
import matplotlib.pyplot as plt

realtime_data = pd.read_csv('./MuJoCo-Realtime-factor.csv', index_col=0)
```

```python
realtime_data
```

```python
num_disks = realtime_data.index.to_numpy()
num_disks
```

```python
dt0 = realtime_data.iloc[:, 0].to_numpy()
dt1 = realtime_data.iloc[:, 1].to_numpy()
dt2 = realtime_data.iloc[:, 2].to_numpy()
dt3 = realtime_data.iloc[:, 3].to_numpy()
dt4 = realtime_data.iloc[:, 4].to_numpy()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(num_disks, dt0, label='dt=0.0001')
plt.plot(num_disks, dt1, label='dt=0.0005')
plt.plot(num_disks, dt2, label='dt=0.001')
plt.plot(num_disks, dt3, label='dt=0.005')
plt.plot(num_disks, dt4, label='dt=0.01')

plt.xlabel('Number of Disks', fontsize=12)
plt.ylabel('Realtime Factor', fontsize=12)
plt.grid(True, which="both")

# Increase tick label sizes
plt.tick_params(axis='both', which='both', labelsize=12)


# Shade the area above y=1
plt.ylim(0.01, 1000)
plt.xlim(0,35)
plt.axhspan(1, plt.ylim()[1], color='gray', alpha=0.3)


plt.legend(fontsize=12)
# plt.xscale('log')
plt.yscale('log')
plt.savefig('mujoco_realtime_factor.png', dpi=300)
plt.show()
```
