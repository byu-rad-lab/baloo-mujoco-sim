```python
import sympy as sp

a1, a2, l1, l2, ml1, ml2, Il1, Il2, kr1, kr2, mm1, mm2, Im1, Im2 = sp.symbols('a_1 a_2 l_1 l_2 m_{l_1} m_{l_2} I_{l_1} I_{l_2} k_{r_1} k_{r_2} m_{m_1} m_{m_2} I_{m_1} I_{m_2}')
vartheta1, vartheta2, dvartheta1, dvartheta2, ddvartheta1, ddvartheta2 = sp.symbols(r'\vartheta_1 \vartheta_2 \dot{\vartheta}_1 \dot{\vartheta}_2 \ddot{\vartheta}_1 \ddot{\vartheta}_2')
```

```python
b11 = Il1 + ml1 * l1**2 + kr1**2 * Im1 + Il2 + ml2 * (
    a1**2 + l2**2 + 2 * a1 * l2 *
    sp.cos(vartheta2)) + Im2 + mm2 * a1**2

b12 = b21 = Il2 + ml2 * (
    l2**2 +
    a1 * l2 * sp.cos(vartheta2)) + kr2 * Im2
b22 = Il2 + ml2 * l2**2 + kr2**2 * Im2

B = sp.Matrix([[b11, b12], [b21, b22]])
B
```

```python
h = -ml2 * a1 * l2 * sp.sin(vartheta2)

c11 = h * dvartheta2
c12 = h * (dvartheta1 + dvartheta2)
c21 = -h * dvartheta1
c22 = 0

C = sp.Matrix([[c11, c12], [c21, c22]])
C
```

```python
g = sp.symbols('g')

g1 = (ml1 * l1 + mm2 * a1 +
        ml2 * a1) * g * sp.cos(
            vartheta1) + ml2 * l2 * g * sp.cos(vartheta1 + vartheta2)
g2 = ml2 * l2 * g * sp.cos(vartheta1 + vartheta2)

G = sp.Matrix([[g1], [g2]])
G
```

```python
xdd1_ref, xdd2_ref = sp.symbols(r'\ddot{x}_1^{\text{ref}} \ddot{x}_2^{\text{ref}}')
xd1_ref, xd2_ref = sp.symbols(r'\dot{x}_1^{\text{ref}} \dot{x}_2^{\text{ref}}')

xdd_ref = sp.Matrix([[xdd1_ref], [xdd2_ref]])
xd_ref = sp.Matrix([[xd1_ref], [xd2_ref]])
```

```python
tau_ref = sp.expand(B @ xdd_ref + C @ xd_ref + G)
```

```python
pi1 = ml1 + mm2
pi2 = ml1 * (l1 - a1)
pi3 = Il1 + ml1 * (l1 - a1)**2 + Im2
pi4 = Im1
pi5 = ml2
pi6 = ml2 * (l2 - a2)
pi7 = Il2 + ml2 * (l2 - a2)**2
pi8 = Im2

pi = sp.Matrix([[pi1, pi2, pi3, pi4, pi5, pi6, pi7, pi8]]).T
pi
```

```python
y11 = a1**2 * ddvartheta1 + a1 * g * sp.cos(vartheta1)
y12 = 2 * a1 * ddvartheta1 + g * sp.cos(vartheta1)
y13 = ddvartheta1
y14 = kr1**2 * ddvartheta1
y15 = ((a1**2 + 2 * a1 * a2 * sp.cos(vartheta2) +
        a2**2) * ddvartheta1 +
        (a1 * a2 * sp.cos(vartheta2) + a2**2) *
        ddvartheta2 - 2 * a1 * a2 * sp.sin(vartheta2) *
        dvartheta1 * dvartheta2 -
        a1 * a2 * sp.sin(vartheta2) * dvartheta2**2 +
        g * a1 * sp.cos(vartheta1) +
        a2 * g * sp.cos(vartheta1 + vartheta2))
y16 = (
    (2 * a1 * sp.cos(vartheta2) + 2 * a2) * ddvartheta1 +
    (a1 * sp.cos(vartheta2) + 2 * a2) * ddvartheta2 -
    2 * a1 * sp.sin(vartheta2) * dvartheta1 * dvartheta2 -
    a1 * sp.sin(vartheta2) * dvartheta2**2 +
    g * sp.cos(vartheta1 + vartheta2))

y17 = ddvartheta1 + ddvartheta2
y18 = kr2 * ddvartheta2
y21 = 0
y22 = 0
y23 = 0
y24 = 0
y25 = ((a1 * a2 * sp.cos(vartheta2) + a2**2) *
        ddvartheta1 + a2**2 * ddvartheta2 +
        a1 * a2 * sp.sin(vartheta2) * dvartheta1**2 +
        g * a2 * sp.cos(vartheta1 + vartheta2))
y26 = ((a1 * sp.cos(vartheta2) + 2 * a2) * ddvartheta1 +
        2 * a2 * ddvartheta2 +
        a1 * sp.sin(vartheta2) * dvartheta1**2 +
        g * sp.cos(vartheta1 + vartheta2))
y27 = ddvartheta1 + ddvartheta2
y28 = kr2 * ddvartheta1 + kr2**2 * ddvartheta2

Y = sp.Matrix([[y11, y12, y13, y14, y15, y16, y17, y18],
                [y21, y22, y23, y24, y25, y26, y27, y28]])

Y
```

```python
#make sure the book is not lying
qdd = sp.Matrix([[ddvartheta1], [ddvartheta2]])
qd = sp.Matrix([[dvartheta1], [dvartheta2]])
sp.simplify(Y @ pi - (B @ qdd + C @ qd + G))
```
